#Question 1: What is a random variable in probability theory?

A random variable is a numerical outcome of a random experiment. It assigns a real number to each possible outcome of a random process.

Example:

Tossing a coin:

Heads = 1

Tails = 0

Here, the value depends on chance, so it is a random variable.

#Question 4: What is a binomial distribution, and how is it used in probability?

A binomial distribution represents the probability of getting a fixed number of successes in a fixed number of independent trials, where:

Each trial has two outcomes (success or failure)

Probability of success is constant

Example uses:

Number of heads in coin tosses

Number of defective products in a batch

#Question 5: What is the standard normal distribution, and why is it important?

#Question 6: What is the Central Limit Theorem (CLT), and why is it critical in statistics?
The Central Limit Theorem (CLT) states that:

As the sample size increases, the sampling distribution of the sample mean approaches a normal distribution, regardless of the population’s distribution.

Why it is critical:

Allows statistical inference

Makes confidence intervals and hypothesis testing possible

Works even if population data is not normally distributed

#Question 7: What is the significance of confidence intervals in statistical analysis?

A confidence interval (CI) gives a range of values within which the true population parameter is expected to lie.

Example:
A 95% CI means we are 95% confident that the true mean lies in that interval.

Importance:


1.   Shows reliability of estimates
1.   Better than single-value estimates
2.   Used widely in decision-making

#Question 9: Python program – Normal Distribution

**Task**

1.   Generate 1000 random numbers
1.   Mean = 50
2.   Standard deviation = 5
2.   Compute mean & std
1.   Plot histogram

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random numbers
data = np.random.normal(50, 5, 1000)

# Calculate mean and standard deviation
mean_value = np.mean(data)
std_value = np.std(data)

print("Mean:", mean_value)
print("Standard Deviation:", std_value)

# Plot histogram
plt.hist(data, bins=30)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Normal Distribution (Mean=50, Std=5)")
plt.show()
```

#Question 10: Sales Trend Analysis using CLT

```python
import numpy as np
import scipy.stats as stats

daily_sales = [120, 130, 125, 140, 135, 128, 150, 145, 138, 155, 160, 152, 148, 165, 170, 162, 158, 175, 180, 172, 168, 185, 190, 182, 178, 195, 200, 192, 188, 205]
sales = np.array(daily_sales)

# Sample statistics
mean_sales = np.mean(sales)
std_sales = np.std(sales, ddof=1)
n = len(sales)

# 95% Confidence Interval
confidence_level = 0.95
z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)
margin_error = z_score * (std_sales / np.sqrt(n))

lower_ci = mean_sales - margin_error
upper_ci = mean_sales + margin_error

print("Mean Sales:", mean_sales)
print("95% Confidence Interval:", (lower_ci, upper_ci))
```
